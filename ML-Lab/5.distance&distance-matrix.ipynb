```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import euclidean_distances
```

```python
def euclidean_distance(x, y):
    return np.linalg.norm(x - y)

def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

x1 = np.array(list(map(float, input("Enter coordinates for point 1 (x1, y1, z1): ").split())))
x2 = np.array(list(map(float, input("Enter coordinates for point 2 (x2, y2, z2): ").split())))

euclidean_dist = euclidean_distance(x1, x2)
manhattan_dist = manhattan_distance(x1, x2)

print("Euclidean Distance:", euclidean_dist)
print("Manhattan Distance:", manhattan_dist)
```

```python
iris = load_iris()
X = iris.data

distance_matrix = euclidean_distances(X)

print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
class1_indices = np.where(iris.target == 0)[0]
class2_indices = np.where(iris.target == 1)[0]

plt.scatter(X[class1_indices, 0], X[class1_indices, 1], label="Class 1")
plt.scatter(X[class2_indices, 0], X[class2_indices, 1], label="Class 2")

m = -0.5  # Slope
c = 1.5  # Intercept

x_line = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_line = m * x_line + c
plt.plot(x_line, y_line, color='red')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
