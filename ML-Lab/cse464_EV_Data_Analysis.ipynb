```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('title_transactions.csv')
df.head()
```

```python
df.info()
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors='coerce')
```

```python
df['month'] = df['transaction_date'].dt.month
df['quartile'] = df['transaction_date'].dt.quarter
```

```python
top_10 = df['county'].value_counts().nlargest(10).index
```

```python
df_top = df[df['county'].isin(top_10)]
```

```python
df_brand_year = df_top.groupby(['county', 'make', 'model_year']).size().reset_index(name='vehicle_count')
df_brand_year.to_csv('counties_brand_year.csv', index=False)
print(df_brand_year.info())
df_brand_year.head()
```

```python
df_brand_quartile_year = df_top.groupby(['county', 'make', 'quartile', 'model_year']).size().reset_index(name='vehicle_count')
df_brand_quartile_year.to_csv('counties_brand_quartile_year.csv', index=False)
print(df_brand_quartile_year.info())
df_brand_quartile_year.head()
```

```python
df_brand_month_year = df_top.groupby(['county', 'make', 'month', 'model_year']).size().reset_index(name='vehicle_count')
df_brand_month_year.to_csv('counties_brand_month_year.csv', index=False)
print(df_brand_month_year.info())
df_brand_month_year.head()
```
