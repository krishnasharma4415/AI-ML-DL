```python
import numpy as np
import pandas as pd
from metapub import PubMedFetcher
```

```python
categories = ["Cancer", "AI in Medicine", "Cardiology", "Neurology", "COVID-19"]
num_of_articles = 10
```

```python
fetch = PubMedFetcher()
```

```python
category_pmids = {} 
for category in categories:
    pmid_list = fetch.pmids_for_query(category, retmax=num_of_articles)
    category_pmids[category] = pmid_list 
```

```python
all_data = []
for category, pmid_list in category_pmids.items():
    for pmid in pmid_list:
        article = fetch.article_by_pmid(pmid)
        title = article.title if article.title else "N/A"
        authors = ", ".join(article.authors) if article.authors else "N/A"
        year = article.year if article.year else "N/A"
        abstract = article.abstract if article.abstract else "N/A"
        
        title_length = len(title)
        abstract_length = len(abstract) if abstract != "N/A" else 0

        article_data = {
            "Category": category,
            "PMID": pmid,
            "Title": title,
            "Authors": authors,
            "Year": year,
            "Abstract": abstract,
            "Title Length": title_length,
            "Abstract Length": abstract_length,
            "Link": f"https://pubmed.ncbi.nlm.nih.gov/{pmid}/"
        }

        all_data.append(article_data)
```

```python
df = pd.DataFrame(all_data)
df.to_csv("pubmed_articles.csv", index=False)
df.head(10)
```
