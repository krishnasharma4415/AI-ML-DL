```python
def log_transform(data, cols):
    for colname in cols:
        if (data[colname] <= 0).any():
            data[colname + '_log'] = np.log1p(data[colname])
            print(f"Column '{colname}' has non-positive values. Using log1p.")
        else:
            data[colname + '_log'] = np.log(data[colname])
            print(f"Column '{colname}' has only positive values. Using log.")

    data.info()
```

```python
def clean_and_convert(df, col_name):
    df[col_name] = df[col_name].astype(str).str.extract('(\d+\.\d+|\d+)')
    df[col_name] = pd.to_numeric(df[col_name], errors='coerce')
    return df

df = clean_and_convert(df, 'Mileage')
df = clean_and_convert(df, 'Engine')
df = clean_and_convert(df, 'Power')
```

```python
df['Seats'].fillna(value=np.nan, inplace=True)  
df['Seats'] = df.groupby(['Model', 'Brand'])['Seats'].transform(lambda x: x.fillna(x.median()))

df['Mileage'].fillna(value=df['Mileage'].mean(), inplace=True)

df['Engine'].fillna(df.groupby(['Brand', 'Model'])['Engine'].transform(lambda x: x.fillna(x.median())), inplace=True)
df['Power'].fillna(df.groupby(['Brand', 'Model'])['Power'].transform(lambda x: x.fillna(x.median())), inplace=True)

df.isnull().sum()
```
