```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
X = np.array([[2, 1], [3, 4], [5, 0], [7, 6], [9, 2]])

mean = np.mean(X, axis=0)

X_centered = X - mean

cov_matrix = np.cov(X_centered.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]


X_transformed = X_centered @ eigenvectors[:, 0:1]

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("Original Feature 1")
plt.ylabel("Original Feature 2")
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.scatter(X_transformed[:, 0], np.zeros_like(X_transformed[:, 0]))
plt.xlabel("Transformed Feature 1")
plt.ylabel("Transformed Feature 2")
plt.title("Transformed Data (1D)")
```

```python
iris = load_iris()
X = iris.data
y = iris.target

mean = np.mean(X, axis=0)

X_centered = X - mean

cov_matrix = np.cov(X_centered.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

X_transformed = X_centered @ eigenvectors[:, 0:2]
```

```python
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y)
plt.xlabel("Transformed Feature 1")
plt.ylabel("Transformed Feature 2")
plt.title("Iris Dataset in 2D Transformed Space")
plt.show()
```
